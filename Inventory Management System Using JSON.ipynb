```python
# Importing Libraries
import json
import time

# Importing Inventory data from Record.json file
fd = open("Record.json","r")
js = fd.read()
fd.close()

# Converting String data to Dictionary
record = json.loads(js)

# Displaying Menu
print("---------------------MENU------------------------\n")
for key in record.keys():
    print(key,":",record[key]["Name"],"| ",record[key]["Price"]," | ",record[key]["Qn"])
print("\n--------------------------------------\n")

# Taking Inputs from the user about their details and purchase

ui_name = str(input("Enter your name         : "))
ui_mail = str(input("Enter Mail ID           : "))
ui_ph    = str(input("Enter Phone No         :  "))
ui_pr_id = str(input("Enter Product ID       : "))
ui_pr_Qn = int(input("Enter Product Quantity : "))

# If we're having equal or more quantity then the user wants

if (ui_pr_Qn <= record[ui_pr_id]["Qn"]):

    print("\n-----------------------------------------------\n")
    print("                   Bill                         \n")

    print("Product Name       :   ",record[ui_pr_id]["Name"])
    print("Product Price (Rs) :   ",record[ui_pr_id]["Price"],"Rs")
    print("Product Quantity   :   ",ui_pr_Qn)


    print("\n-----------------------------------------------\n")

    print("Billing Amount :",(record[ui_pr_id]["Price"])*ui_pr_Qn,"Rs")

    print("\n-----------------------------------------------\n")

     # Generating CSV Transection Detail
    
    sale = '1'+","+ui_name+","+ui_mail+","+ui_ph+","+ui_pr_id+","+record[ui_pr_id]["Name"]+","+str(ui_pr_Qn)+","+str(record[ui_pr_id]["Price"])+","+str(ui_pr_Qn * record[ui_pr_id]["Price"])+","+time.ctime()+"\n"
        
     # Updating Inventory in Dictionary
    record[ui_pr_id]['Qn'] = record[ui_pr_id]['Qn'] - ui_pr_Qn
    
# If we're less quantity then the user wants    
else:
    print("Sorry We are not having enough Quantity in our Inventory\n")
    print("We are only having",record[ui_pr_id]['Qn'],"Only")
    print("\n-----------------------------------------------\n")
    ch = input("Press Y/y to Purchase :")
    
     # If user wants to purchase the whole quantity for that product
    if ch == "Y" or ch == "y":
    
        print("\n-----------------------------------------------\n")
        print("                   Bill                         \n")

        print("Product Name       :   ",record[ui_pr_id]["Name"])
        print("Product Price (Rs) :   ",record[ui_pr_id]["Price"],"Rs")
        print("Product Quantity   :   ",record[ui_pr_id]["Qn"])


        print("\n-----------------------------------------------\n")

        print("Billing Amount :",(record[ui_pr_id]["Price"])*record[ui_pr_id]["Qn"],"Rs")

        print("\n-----------------------------------------------\n")
        
        # Generating CSV Transection Detail
        sale = '1'+","+ui_name+","+ui_mail+","+ui_ph+","+ui_pr_id+","+record[ui_pr_id]["Name"]+","+str(record[ui_pr_id]['Qn'])+","+str(record[ui_pr_id]["Price"])+","+str(record[ui_pr_id]['Qn'] * record[ui_pr_id]["Price"])+","+time.ctime()+"\n"
        
        # Updating Inventory in Dictionary
        
        record[ui_pr_id]['Qn'] = 0
        
    # If user pressed anything except Y or y
    else:
        print("\nThank You")

    

# Converting Inventory Dictionary to String
js = json.dumps(record)

# Updating Inventory and Saving in to my Records.json
fd = open('Record.json','w')
fd.write(js)
fd.close()

# Adding Transection on Sales File
fd = open("sales.txt","a")
fd.write(sale)
fd.close()


print('')
print("---------------------------------------------")
print("  Thanks for your order, Inventory Updated!  ")
print("---------------------------------------------")
```
